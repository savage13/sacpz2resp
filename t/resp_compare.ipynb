```python
amppz = np.loadtxt('AMP.IU.ANMO.00.BH1.pz')
phapz = np.loadtxt('PHASE.IU.ANMO.00.BH1.pz')
amp = np.loadtxt('AMP.IU.ANMO.00.BH1')
pha = np.loadtxt('PHASE.IU.ANMO.00.BH1')
```
